1Ô∏è‚É£ Importing Libraries

```python
import numpy as np
import pandas as pd
```

2Ô∏è‚É£ Loading the Dataset

```python
df = pd.read_csv("coffee_shop_sales.csv")
```

3Ô∏è‚É£ Converting Data Types (Pandas)

```python
df['Date'] = pd.to_datetime(df['Date'])
```

4Ô∏è‚É£ Checking & Handling Missing Values

```python
df.isnull().sum()        # Check missing values
df = df.dropna()         # Remove rows with missing values
```

5Ô∏è‚É£ Adding New Columns (NumPy + Pandas)
‚úî Extracting Month & Day

```python
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day_name()

# Creating High/Low Sales Flag (NumPy)


df['Sales_Tag'] = np.where(df['Total_Sales'] > 10, 'High', 'Low')
```

6Ô∏è‚É£ Grouping & Aggregation

```python
# Total Sales per Category
category_sales = df.groupby('Category')['Total_Sales'].sum()

# Total Sales per Month
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total_Sales'].sum()

```

7Ô∏è‚É£ Sorting Data

```python
df.sort_values(by='Total_Sales', ascending=False).head()
```

8Ô∏è‚É£ Filtering Data

```python
#Filter all Coffee category sales

df[df['Category'] == 'Coffee']

#Filter Morning sales
df[df['Time_Period'] == 'Morning']
```

9Ô∏è‚É£ Using NumPy for Calculations

```python
average_sale = np.mean(df['Total_Sales'])
max_sale     = np.max(df['Total_Sales'])
min_sale     = np.min(df['Total_Sales'])
```
